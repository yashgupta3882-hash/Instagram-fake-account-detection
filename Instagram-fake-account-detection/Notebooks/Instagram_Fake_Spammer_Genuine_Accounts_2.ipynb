```
#Countplot
sns.countplot(x='fake', data=df2)
plt.title('Distribution of Fake vs Genuine Accounts')
plt.show()
```

```
#Correlation
correlation = df2.corr()
plt.figure(figsize=(12,10))
sns.heatmap(correlation, annot = True, cmap= 'coolwarm')
plt.title('Feature Correlation Matrix')
plt.show()
```

```
#Barplot
sns.barplot(x='fake', y='profile pic', data = df2)
plt.title('Profile Picture Presence in Fake vs Genuine Accounts')
plt.show()
```

```
#Boxplot
sns.boxplot(x='fake', y='#followers', data=df2)
plt.title('Followers Count in Fake vs Genuine Accounts')
plt.show()
```

```
#Boxplot
sns.boxplot(x='fake', y='#follows', data=df2)
plt.title('Following Count in Fake vs Genuine Accounts')
plt.show()
```

```
#Boxplot
sns.boxplot(x='fake', y='#posts', data=df2)
plt.title('Posts Count in Fake vs Genuine Accounts')
plt.show()
```

```
#Histogram
df2.hist(figsize=(12,10))
plt.show()
```

```
#Countplot
plt.figure(figsize=(12,10))
sns.countplot(x='#followers', hue = 'fake', data = df2)
plt.show()
```

```
plt.figure(figsize=(20,10))
sns.countplot(x='private', hue='fake', data=df2)
plt.title('Fake Distributed among private accounts')
plt.show()
```
